```python
# 1. Generate M x N random noise image.
# 1.1 uniform noise
def random_noise(M, N, type):
  if type == "uniform":
    image = np.random.uniform(0, 1, size = (M, N))
    image = image * 255
    return image
  elif type == "gaussian":
    image = np.random.normal(0, 1, size = (M, N))
# 아래 식 수정 필요
    image = 0.5 + image / (2 * np.max(image))
    image = image * 255
    return image
M = 512
N = 512
image1 = random_noise(M, N, type='uniform')
plt.subplot(2, 2, 1)
plt.title('random noise, uniform')
plt.imshow(image1, cmap="gray")
# 1.1 histogram
histo = np.zeros(256)
for p in range(256):
  histo[p] = np.count_nonzero(np.round(image1)==p)
plt.subplot(2, 2, 2)
plt.bar(np.arange(256),histo)
plt.title('histogram')
# 1.2 Gaussian noise
image2 = random_noise(M, N, type='gaussian')
plt.subplot(2, 2, 3)
plt.title('random noise, gaussian')
plt.imshow(image2, cmap="gray")
# 1.2 histogram
histo = np.zeros(256)
for p in range(256):
  histo[p] = np.count_nonzero(np.round(image2)==p)
plt.subplot(2, 2, 4)
plt.bar(np.arange(256),histo)
plt.title('histogram')
plt.show
```
