```python
# 1. RGB to YUV function
def RGB2YUV(img, a=1):
  A = [[0.299, 0.587, 0.114], [-0.147, -0.289, 0.436], [0.615, -0.515, -0.100]]
  yuv = np.zeros(img.shape)
  if a==-1:
    for i in range(img.shape[0]):
      for j in range(img.shape[1]):
        yuv[i,j,:] = np.transpose(np.matmul(np.linalg.inv(A),img[i,j,:]))
  else:
    for i in range(img.shape[0]):
      for j in range(img.shape[1]):
        yuv[i,j,:] = np.transpose(np.matmul(A,img[i,j,:]))
  yuv.astype(int)
  return yuv
```
