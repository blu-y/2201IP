```python
import numpy as np
import matplotlib.pyplot as plt
import math
import cv2
image = cv2.imread('/content/drive/My Drive/Colab Notebooks/lena.bmp',cv2.IMREAD_GRAYSCALE)
```

```python
# 1. Design image 2x scaling function.
# 임의의 배수로 할 줄 알아야 함.
# 반으로 줄이고 다시 스케일링하여 확인 가능
# Forward 말고 Backward로 짜라 - 정수가 아닐 수도 있음
def scaling(image, n):
   result = np.zeros([image.shape[0]*n,image.shape[1]*n],dtype=np.uint8)
   image_ = np.zeros([image.shape[0]+1,image.shape[1]+1],dtype=np.uint8)
   image_[0:image.shape[0],0:image.shape[1]] = image[:,:]
   image_[image.shape[0], image.shape[1]] = image[image.shape[0]-1, image.shape[1]-1]
   for i in range(image.shape[0]*n):
     for j in range(image.shape[1]*n):
       b = (i%n)/n
       a = (j%n)/n
       x = int(i/n)
       y = int(j/n)
       result[i,j] = (1-a)*((1-b)*image_[x,y]+b*image_[x,y+1])+a*((1-b)*image_[x+1,y]+b*image_[x+1,y+1])
   return result
result = scaling(image,3)
plt.title('Scaling')
plt.imshow(result,cmap="gray")
cv2.imwrite('/content/drive/My Drive/Colab Notebooks/lena_scale3.bmp', result)
print(result.shape)
```
