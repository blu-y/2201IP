```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('/content/drive/My Drive/Colab Notebooks/image.jpg')

def filter_colors(img, gray=0):
    img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    w_mask = cv2.inRange(img, (200, 200, 200), (255, 255, 255))
    y_mask = cv2.inRange(img_hsv, (10, 100, 100), (40, 255, 255))
    w_img = np.zeros(np.shape(img))
    y_img = np.zeros(np.shape(img))
    w_img = cv2.bitwise_and(img, img, w_img, w_mask)
    y_img = cv2.bitwise_and(img_hsv, img_hsv, y_img, y_mask)
    img_f = cv2.addWeighted(w_img, 1.0, y_img, 1.0, 0.0)
#    plt.imshow(cv2.cvtColor(img_f, cv2.COLOR_BGR2RGB))
#    plt.show()
    if gray == 1:
        img_f = cv2.cvtColor(img_f, cv2.COLOR_BGR2GRAY)
    return img_f

def limit_region(img):
    w_img = np.zeros(np.shape(img))
    output = img
    return output

img_f = filter_colors(img, gray=1)
img_edges = np.zeros(np.shape(img_f))
img_edges = cv2.Canny(img_f, 50, 150)

plt.imshow(img_f, cmap='gray')
plt.show()
plt.imshow(img_edges, cmap='gray')
plt.show()

img_mask = limit_region(img_edges)
#asd
```
