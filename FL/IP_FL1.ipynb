```python
import numpy as np
import matplotlib.pyplot as plt
import math
import cv2
image = cv2.imread('/content/drive/My Drive/Colab Notebooks/lena.bmp')
rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.title('original')
plt.imshow(rgb)
```

```python
# 1. Convert RGB color image to gray.
# Y = 0.3 R + 0.5 G + 0.2 B
R=rgb.shape[0]
C=rgb.shape[1]
gray = np.zeros([rgb.shape[0], rgb.shape[1]],dtype=np.uint8)
for r in range(R):
 for c in range(C):
   gray[r,c] = 0.3 * rgb[r,c,0] + 0.5 * rgb[r,c,1] + 0.2 * rgb[r,c,2]
plt.title('gray')
plt.imshow(gray,cmap="gray")
```

```python
# 2. Change the input image intensity
# Input : image, offset / Output : image
offset = 50
inten = np.zeros([rgb.shape[0], rgb.shape[1], 3],dtype=np.uint8)
for r in range(R):
  for c in range(C):
    for p in range(3):
      inten[r,c,p] = np.clip(rgb[r,c,p]+offset,0,255)
plt.title('intensity')
plt.imshow(inten)
```

```python
# 3. Find the histogram for given input gray image.
histo = np.zeros(256)
for p in range(256):
  histo[p] = np.count_nonzero(gray==p)
plt.bar(np.arange(256),histo)
plt.title('histogram')
plt.show
```
